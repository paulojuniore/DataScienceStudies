```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional
plt.plot(x, y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('TÃ­tulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r*-')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g--')
```

```python
# Orientada a objetos
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('TÃ­tulo do GrÃ¡fico')
axes2.set_title('TÃ­tulo do GrÃ¡fico 2')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'm-.')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('TÃ­tulo')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('TÃ­tulo do grÃ¡fico')
```

```python
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(5, 5))
axes.plot(x, y, 'r--', label='x ** 2')
axes.plot(y, x, 'b--', label='x ** 1/2')
axes.set_title('ParÃ¡bola')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.legend()
```
